```python
# 2. Exercício: Verificar se há valores NaN nas colunas do DataFrame.

import pandas as pd

df = pd.read_excel('dados_ausentes.xlsx')
display(df.isnull())

```

```python
 # 3. Exercício: Substituir os valores NaN de uma coluna por 0.

df['Salário'].fillna(0, inplace=True)
display(df)
```

```python
# 4. Exercício: Substituir os valores NaN de uma coluna por um valor específico (por exemplo, a média).

df['Idade'].fillna(df['Idade'].mean(), inplace=True)
display(df)
```

```python
# 5. Exercício: Remover todas as linhas que contêm valores NaN em qualquer coluna

df_sem_nulos = df.dropna()
display(df_sem_nulos)
```

```python
# 6. Exercício: Remover todas as linhas que contêm valores NaN em uma coluna específica.

df_remover_cidade = df.dropna(subset=['Cidade'])
display(df_remover_cidade)
```

```python
# 7. Exercício: Substituir valores NaN de uma coluna com a mediana dessa coluna.
df2 = pd.read_excel("dados_ausentes.xlsx")

df2['Salário'].fillna(df['Salário'].median(), inplace=True)
display(df2)
```

```python
# 8. Exercício: Substituir valores NaN de uma coluna com o valor de uma outra coluna.

df3 = pd.read_excel("dados_ausentes.xlsx")

df3['Salário'].fillna(df['Idade'].median(), inplace=True)
display(df3)
```

```python
# 9. Exercício: Preencher valores NaN com o valor anterior (método de &quot;forward fill&quot;).

df4 = pd.read_excel("dados_ausentes.xlsx")

df4['Idade'] = df['Idade'].ffill()
display(df4)
```

```python
# 10. Exercício: Preencher valores NaN com o valor posterior (método de &quot;backward fill&quot;) 

df5 = pd.read_excel("dados_ausentes.xlsx")

df5['Idade'] = df['Idade'].bfill()
display(df5)
```

```python
# 11. Exercício: Alterar todos os valores de uma coluna para maiúsculas.

df['Cidade'] = df['Cidade'].str.title()
display(df)
```

```python
# 12. Exercício: Alterar todos os valores de uma coluna para minúsculas.

df6 = pd.read_excel('dados_ausentes.xlsx')

df6['Nome'] = df['Nome'].str.casefold()
display(df6)
```

```python
# 13. Exercício: Remover espaços em branco à esquerda e à direita em uma coluna.

df7 = pd.read_excel('dados_ausentes.xlsx')

df7['Nome'] = df['Nome'].str.strip()

display(df7)


```

```python
# 14. Exercício: Substituir valores de uma coluna com base em um dicionário de mapeamento (ex: &quot;São Paulo&quot; para &quot;SP&quot;).

df['Cidade'].str.lower().replace({'são paulo': 'SP', 'rio de janeiro': 'RJ', 'rj': 'RJ', 'sp': 'SP'})
display(df)
```

```python
# 15 - Exercício: Identificar e remover valores duplicados de um DataFrame.

df = df.drop_duplicates(subset=['Nome'])
display(df)
```
